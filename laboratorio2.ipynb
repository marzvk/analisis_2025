# Siniestros por zona

```python
# eventos
conteo_por_zona = df.groupby(['Accident_Severity', 'Local_Authority_(District)']).size().reset_index(name='cantidad_eventos')
conteo_por_zona
```

```python
top_20_distritos = (
    conteo_por_zona
    .groupby('Local_Authority_(District)')['cantidad_eventos']
    .sum()
    .sort_values(ascending=False)
    .head(20)
    .index
)
top_20_distritos
```

```python
top_20 = conteo_por_zona[conteo_por_zona['Local_Authority_(District)'].isin(top_20_distritos)]

sns.lineplot(
    data=top_20,
    x='Local_Authority_(District)',
    y='cantidad_eventos',
    hue='Accident_Severity',
    marker='o'
)
plt.xticks(rotation=45, ha='right')
plt.title('Top 20 distritos por cantidad de accidentes, divididos por gravedad')
plt.xlabel('Distrito')
plt.ylabel('Cantidad de eventos')
plt.tight_layout()
plt.show()
```

```python
pivot = top_20.pivot_table(
    index='Local_Authority_(District)',
    columns='Accident_Severity',
    values='cantidad_eventos',
    fill_value=0
).astype(int)

plt.figure(figsize=(10, 8))
sns.heatmap(pivot, annot=True,fmt="d", cmap="YlOrRd")
plt.title('Cantidad de accidentes por distrito y gravedad')
plt.xlabel('Gravedad')
plt.ylabel('Distrito')
plt.tight_layout()
plt.show()
```

# Analisis de siniestros dada la severidad/tipo de vehiculo

```python
conteo_por_tipo_vehiculo = df.groupby(['Accident_Severity', 'Vehicle_Type']).size().reset_index(name='cantidad_eventos')
df_agrupado = conteo_por_tipo_vehiculo.copy()

# Agrupar todos los tipos de motos bajo "Motorcycle"
df_agrupado['Vehicle_Type'] = df_agrupado['Vehicle_Type'].replace({
    vt: 'Motorcycle' for vt in df_agrupado['Vehicle_Type'].unique() if 'Motorcycle' in vt
})

# Agrupar por severidad y tipo de vehículo (motos ya están fusionadas)
df_motos_agrupadas = df_agrupado.groupby(['Accident_Severity', 'Vehicle_Type'])['cantidad_eventos'].sum().reset_index()

grupo = df_motos_agrupadas.groupby(['Vehicle_Type', 'Accident_Severity'])['cantidad_eventos'].sum().reset_index()

# Sumar total de eventos por tipo de vehículo
top_vehiculos = grupo.groupby('Vehicle_Type')['cantidad_eventos'].sum().reset_index()

# Obtener los 5 con más eventos
top_5 = top_vehiculos.sort_values(by='cantidad_eventos', ascending=False).head(5)['Vehicle_Type'].tolist()
# Filtrar solo top 5
grupo_top5 = grupo[grupo['Vehicle_Type'].isin(top_5)]

# Calcular % de cada severidad sobre total por vehículo
grupo_top5['porcentaje'] = grupo_top5.groupby('Vehicle_Type')['cantidad_eventos'].transform(lambda x: x / x.sum() * 100)
grupo_top5
```

```python


pivot = grupo_top5.pivot(index='Vehicle_Type', columns='Accident_Severity', values='porcentaje')


pivot.plot(kind='bar', stacked=True, colormap='Set2', figsize=(10,6))

plt.title('Porcentaje de severidad por tipo de vehículo (Top 5)')
plt.ylabel('Porcentaje (%)')
plt.xlabel('Tipo de vehículo')
plt.xticks(rotation=45)
plt.legend(title='Accident Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 7ma Conclusion
# ✅ Severidad vs tipo vehiculo
Se observan
distribuciones demasiado uniformes o limpias,
porcentajes por severidad (Fatal, Serious, Slight) son muy similares para todos los tipos de vehículo, eso no es lo habitual en datos reales. En la realidad:

Las motos tienden a tener más accidentes graves o fatales por exposición.

Los autos tienen más volumen, pero con menor proporción fatal.


Ausencia de outliers:
atos reales tienen  excepciones, valores extremos. Si todo está “perfectamente balanceado”, parece ser sospechoso. \
Se evidencia un mismo porcentaje de “Slight” para motos y buses , eso es raro.

# Siniestros por condiciones climaticas

```python
conteo_por_condiciones_climaticas = df.groupby(['Accident_Severity', 'Weather_Conditions']).size().reset_index(name='cantidad_eventos')


# Crear tabla pivote
pivot = conteo_por_condiciones_climaticas.pivot_table(
    index='Weather_Conditions',
    columns='Accident_Severity',
    values='cantidad_eventos',
    fill_value=0
)


pivot_porcentaje = pivot.div(pivot.sum(axis=1), axis=0) * 100

# Visualizar con heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_porcentaje, annot=True, fmt=".1f", cmap="Reds")
plt.title("Porcentaje de Accidentes por Severidad y Condición Climática")
plt.ylabel("Condición Climática")
plt.xlabel("Severidad del Accidente")
plt.tight_layout()
plt.show()
```

# 8va Conclusion
# ✅ Condicion climatica/ severidad
Se evidencia que los accidentes fatales se incrementan bajo condiciones :

*   Fog or mist
* **Llamativamente** no inciden tanto las condiciones :

*  Snowing + high winds
* Rainig + high winds

# Area rural/urbana

```python
# Urban_or_Rural_Area
conteo_por_area_urbana_rural = df.groupby(['Urban_or_Rural_Area','Accident_Severity', ]).size().reset_index(name='cantidad_eventos')
```

```python
tabla_area = conteo_por_area_urbana_rural.pivot_table(
    index='Urban_or_Rural_Area',
    columns='Accident_Severity',
    values='cantidad_eventos',
    fill_value=0
)
tabla_area
```
